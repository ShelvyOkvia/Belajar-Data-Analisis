```python
def Data_Summary(data, col1, col2='TARGET', sort=False):
    temp = pd.DataFrame(application_train.groupby(col1)[col2].agg(lambda x: x.eq(1).sum())).reset_index()

    temp['total'] = pd.DataFrame(application_train.groupby(col1)[col2].agg(total='count')).reset_index()['total']

    if sort == False:
        temp.sort_values(by=['total'],inplace=True, ascending=False)


    print(temp)
```

**Label Encoding and One-Hot Encoding**



```python
from sklearn.preprocessing import LabelEncoder
# Create a label encoder object
le = LabelEncoder()
```

```python
# let's break down the columns by their type (i.e. int64, float64, object)
application_full.dtypes.value_counts()
```

```python
le_count = 0
for col in application_full.columns[1:]:
    if application_full[col].dtype == 'object':
        if len(list(application_full[col].unique())) <= 2:
            le.fit(application_full[col])
            application_full[col] = le.transform(application_full[col])
            le_count += 1
print('{} columns were label encoded.'.format(le_count))
```

```python
# convert rest of categorical variable into dummy
application_full = pd.get_dummies(application_full)
application_full
```
